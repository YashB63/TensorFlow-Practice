```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Make a bigger dataset
X = tf.range(-100, 100, 4) # -100 se le kar 100 aur har number ke bich 4 ka difference
X
```

```python
y = X + 10
y
```

```python
plt.scatter(X,y)
plt.show()
```

```python
len(X)
```

```python
# Split the data into train and test sets
X_train = X[:40] # first 40 are training samples (80% of the data)
y_train = y[:40]

X_test = X[40:] # last 10 are testing samples (20% of the data)
y_test = y[40:]

len(X_train), len(X_test), len(y_train), len(y_test)
```

## Visualizing the Data

#### Now we have got our data in training and test sets .... Lets's visualize again!

```python
plt.figure(figsize=(10,7))

# Plot the training data in magenta
plt.scatter(X_train, y_train, c="m", label="Training Data")

# Plot the test data in cyan
plt.scatter(X_test, y_test, c="cyan", label="Testing Data") 

# Show a legend
plt.legend()

plt.show()
```
